```python
entries_rain = weather_subway[weather_subway['rain']==1].ENTRIESn_hourly
entries_norain = weather_subway[weather_subway['rain']==0].ENTRIESn_hourly
```

```python
scipy.stats.shapiro(entries_norain)
scipy.stats.shapiro(entries_rain)
```

```python
n_one = len(entries_norain)
n_two = len(entries_rain)

scipy.stats.mannwhitneyu(entries_norain, entries_rain)
```

```python
ggplot(aes(x='ENTRIESn_hourly', color='rain', fill='rain'), data = weather_subway) + \
    geom_density() + \
    xlab("Hourly passenger entries") + \
    ylab("Frequency") + \
    ggtitle("Hourly passenger entries (weekdays and weekends)") + \
    theme_seaborn()
```

```python
def STD(n1, n2):
    return np.sqrt((n1*n2)*(n1+n2+1)/12)
```

```python
STD(n_one, n_two)
```

```python
z_score = (scipy.stats.mannwhitneyu(entries_norain, entries_rain)[0]-(n_one*n_two/2))/STD(n_one, n_two)
print(z_score)
```

```python
p_values = 1 - scipy.stats.norm.sf(z_score)
print(p_values)
print(p_values * 2)
```
